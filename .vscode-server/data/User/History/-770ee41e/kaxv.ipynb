```python
from dotenv import load_dotenv
import os

# Load environment variables from .env
load_dotenv()

# Pull the key
api_key = os.getenv("YELP_API_KEY")

if not api_key:
    raise EnvironmentError("❌ YELP_API_KEY not loaded. Check your .env file and load_dotenv().")

print("✅ Key loaded successfully")
```
